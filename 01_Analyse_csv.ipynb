# Analyse des csv brute 

```python
"""
Importation de la librairie pandas et 
seaborn qui sera utilisée tout au 
long des scripts d'analyse
"""
import pandas as pd
import seaborn as sns
```

### 1 - DatafinitiElectronicsProductData

#### Lecture de table ProductData + affichage :

```python
df1 = pd.read_csv('BDD/Brut_csv/DatafinitiElectronicsProductData.csv')
df1.head(5)
```

#### Aperçus des données :

```python
print(df1.info())
print("\n \n \n")
print(df1.isnull().sum())
print("\n \n")
```

#### Pour aller plus loin :

##### Aperçus graphique des données :

```python
sns.heatmap(df1.isnull(),yticklabels=False, cbar=False,cmap='viridis')
```
